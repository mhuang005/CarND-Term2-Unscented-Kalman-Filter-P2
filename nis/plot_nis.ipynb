# Consistency -- Normalized Innovation Squared (NIS)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Plot laser NIS 
with open('nis_laser.txt', 'r') as f:
    y_laser = f.read().splitlines()
    
x_laser = list(range(1, len(y_laser)+1))
y_laser = list(map(float, y_laser))

    
plt.plot(x_laser, y_laser)
plt.axhline(y=0.103, color='g')
plt.axhline(y=5.991, color='r')
plt.axis([-1, 260, -1, 10])
_ = plt.title('Normalized Innovation Squared (NIS) for Laser Data')
        
```

```python
# Plot radar NIS
with open('nis_radar.txt', 'r') as f:
    y_radar = f.read().splitlines()
    
x_radar = list(range(1, len(y_radar)+1))
y_radar = list(map(float, y_radar))

    
plt.plot(x_radar, y_radar)
plt.axhline(y=0.352, color='g')
plt.axhline(y=7.815, color='r')
plt.axis([-1, 260, -1, 15])
_ = plt.title('Normalized Innovation Squared (NIS) for Radar Data')
        
```
